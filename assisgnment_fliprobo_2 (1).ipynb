# 2)  -In the above question, now scrape the following details of each product listed in first 3 pagesof your search results and save it in a data frame and csv. In case if any product vertical hasless than 3 pages in search results then scrape all the products available under that productvertical. Details to be scraped are: "Brand Name", "Name of the Product", "Rating", "No. of Ratings", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Other Details"and “Product URL”. In case, if any of the details are missing for any of the product thenreplace it by “-“.


```python
#Creating empty list
Brand_Name = []
Name_of_the_Product = []
Rating  = []
Total_Ratings  = []
Price  = []
Return_Exchange  = []
Expected_Delivery  = []
Availability = []
Other_Details = []
Product_URL = []
```

```python
Task1 = driver.find_elements_by_xpath("//span[@class='a-size-base-plus a-color-base']")

for hy in Task1[5:21]:
        se = hy.text
        Brand_Name.append(se)

for gh in Task1[33:64]:
            se1 = gh.text
            Brand_Name.append(se1)
```

```python
start_page=1
end_page=3
    
    
for page in range(start_page,end_page+1):
        
        Task1 = driver.find_elements_by_xpath("//span[@class='a-size-base-plus a-color-base']")

        for hy in Task1[5:21]:
            se = hy.text
            Brand_Name.append(se)
        time.sleep(2)

        for gh in Task1[33:64]:
            se1 = gh.text
            Brand_Name.append(se1)
            
        Task1 = driver.find_elements_by_xpath("//h2[@class='a-size-mini a-spacing-none a-color-base s-line-clamp-2']/a/span")

        for hy in Task1[5:21]:
            se = hy.text
            Name_of_the_Product.append(se)
        time.sleep(3)

        for gh in Task1[33:64]:
            se1 = gh.text
            Name_of_the_Product.append(se1)

        urls1=[i.get_attribute("href")for i in driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']")]
        for yu in urls1[5:21]:
            Product_URL.append(yu)
            
        time.sleep(3)
    
        for ui in urls1[33:64]:
            Product_URL.append(ui)

    
            
        time.sleep(3)
        nxt_button1=driver.find_element_by_xpath("//a[text()='Next']").click()
            
```

```python
print(len(Product_URL))
print(len(Brand_Name))
print(len(Name_of_the_Product))
```

```python
for url in Product_URL:
    driver.get(url)

        
        
    try:
        Task3 = driver.find_element_by_xpath("//div[@id='averageCustomerReviews']/span[3]/a/span").text
    
        Total_Ratings.append(Task3)
    except NoSuchElementException: 
        Total_Ratings.append('--')
        
    try:
        Task3 = driver.find_element_by_xpath("//span[@id ='priceblock_ourprice'] | //span[@id ='priceblock_saleprice'] |  //span[@id ='priceblock_dealprice']").text
    
        Price.append(Task3)
    except NoSuchElementException: 
        Price.append('--')
        
    try:
        Task3 = driver.find_element_by_xpath("//div/span/div[2]/a[@class ='a-size-small a-link-normal a-text-normal']").text
    
        Return_Exchange.append(Task3)
    except NoSuchElementException: 
        Return_Exchange.append('--')
        
    try:
        Task3 = driver.find_element_by_xpath("//div[@id='ddmDeliveryMessage']/b").text
    
        Expected_Delivery.append(Task3)
    except NoSuchElementException: 
        Expected_Delivery.append('--')
        
        
    try:
        Task3 = driver.find_element_by_xpath("//input[@id='add-to-cart-button']")
    
        Availability.append('Available')
    except NoSuchElementException: 
        Availability.append('Not Available')
        
    try:
        Task3 = driver.find_element_by_xpath("//ul[@class='a-unordered-list a-vertical a-spacing-mini']").text
    
        Other_Details.append(Task3)
    except NoSuchElementException: 
        Other_Details.append('--')
        
    try:                  
        rate=driver.find_element_by_xpath("//span[@id='acrCustomerReviewText']")
        rate.click()                                                      
        rating=driver.find_element_by_xpath("//span[@class='a-size-medium a-color-base']")#locating the rating
        Rating.append(rating.text)#appending the ratings in Ratings list
        
    except NoSuchElementException   as e:
        Rating.append("NO rating")#appending the No rating if no rating is there
        
        
```

```python
Sunglasses = pd.DataFrame({'Brand_Name' : Brand_Name[0:99], 'Name of the Product': Name_of_the_Product[0:99] ,'Rating' :Rating[0:99] ,'Total_Ratings' :Total_Ratings[0:99] ,'Price' :Price[0:99] ,'Return_Exchange' :Return_Exchange[0:99] ,'Expected_Delivery': Expected_Delivery[0:99] ,'Availability' :Availability[0:99] , 'Other_Details':Other_Details[0:99] ,'Product_URL': Product_URL[0:99]})
Sunglasses
```
